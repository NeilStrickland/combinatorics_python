**Example 13.27**

```python
T = Tournament.odd_modular_tournament(5)
U = Tournament.odd_modular_tournament(5)
TU = Tournament.colon(T, U)
display(Latex('Scores for $T$ are ' + str(T.sorted_scores)))
display(Latex('Scores for $U$ are ' + str(U.sorted_scores)))
display(Latex('Scores for $(T:U)$ are ' + str(TU.sorted_scores)))
```

**Example 13.31**

```python
T = Tournament.odd_modular_tournament(3)
U = Tournament.consistent_tournament(5)
TU = Tournament.star(T, U)
UT = Tournament.star(U, T)
display(Latex('Scores for $T$ are ' + str(T.sorted_scores)))
display(Latex('Scores for $U$ are ' + str(U.sorted_scores)))
display(Latex('Scores for $(T*U)$ are ' + str(TU.sorted_scores)))
display(Latex('Scores for $(U*T)$ are ' + str(UT.sorted_scores)))
```
