**Problem 7.1**

```python
M = MatchingProblem(['a','b','c'],[1,2,3],[('a',3)]).complement()
print(MatchingProblem.print_condensed(M.relation))
M.show_board(labels=True)
M.rook_report()
```

```python
fig, ax = plt.subplots(1,4,figsize=(16,4))
for i,p in enumerate(M.rook_list_by_size[3]):
    ax[i].set_aspect('equal')
    ax[i].axis('off')
    M.show_board(ax=ax[i],matching=p,coloured=False)
```

**Example 7.7**

```python
F3 = MatchingProblem.full_board(['a','b','c'],[1,2,3])
F3.show_board(labels=True) 
plt.show()
F3.rook_report()
fig, ax = plt.subplots(1,6,figsize=(18,3))
for i,p in enumerate(F3.rook_list_by_size[3]):
    ax[i].set_aspect('equal')
    ax[i].axis('off')
    F3.show_board(ax=ax[i],matching=p,coloured=False)
    ax[i].text(1.5,-3.5,MatchingProblem.print_condensed(tuple(p)),fontsize=10,ha='center')
```
