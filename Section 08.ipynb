**Problem 8.9**

```python
L = np.array([[1,2,3,4,5],[2,4,5,3,1],[4,3,1,5,2]],dtype=int)
sp.Matrix(L)
```

```python
A = ['a','b','c','d','e']
B = [1,2,3,4,5]
E = [(x,y) for (i,x) in enumerate(A) for y in B if y not in L.T[i]]
B1 = MatchingProblem(A,B,E)
B1.show_board(labels=True)
```

```python
s1 = ('b',5)
B9, B2 = B1.block_and_strip(*s1)
s2 = ('c',4)
B6, B3 = B2.block_and_strip(*s2)
s9 = ('c',4)
B13, B10 = B9.block_and_strip(*s9)
F3 = ['d']
B4, B5 = B3.factor(F3)
F6 = ['a','e']
B7, B8 = B6.factor(F6)
F10 = ['a','e']
B11, B12 = B10.factor(F10)
F13 = ['b','c','d']
B14, B15 = B13.factor(F13)
fig, ax = plt.subplots(4,3,figsize=(9,12))
for i in range(4):
    for j in range(3):
        ax[i,j].axis('off')
        ax[i,j].set_aspect('equal')
B1.show_board(ax=ax[0,0],caption='$B_1$',height=5,marker=s1)
B2.show_board(ax=ax[0,1],caption='$B_2$',height=5,marker=s2)
B3.show_board(ax=ax[0,2],caption='$B_3$',height=5,factor=F3,factor_labels=('P','Q'))
B6.show_board(ax=ax[1,1],caption='$B_6$',height=5,factor=F6,factor_labels=('R','S'))
B9.show_board(ax=ax[2,0],caption='$B_9$',height=5,marker=s9)
B10.show_board(ax=ax[2,1],caption='$B_{10}$',height=5,factor=F10,factor_labels=('T','U'))
B13.show_board(ax=ax[3,0],caption='$B_{13}$',height=5,factor=F13,factor_labels=('V','W'))

plt.show()
for i in [4,5,7,8,11,12,14,15,3,6,10,13,9,2,1]:
    Bi = locals()['B'+str(i)]
    display(Latex('$r_{' + str(i) + '}(x)=' + Bi.rook_polynomial_latex + '$'))
```
