**Example 6.6**

```python
P = ['a','b','c','d','e']
J = ['l','m','n','o','p']
E = 'al,am,ap,bl,bn,bo,cl,dm,dn,do,dp,el,eo,ep'
E = {(E[3*i],E[3*i+1]) for i in range((len(E)+1)//3)}
M = MatchingProblem(P,J,E)
M.left_pos = {
 'a' : ( 8.80,-4.50),
 'b' : (12.20,-4.50),
 'c' : (10.50,-6.40),
 'd' : (10.50,-1.40),
 'e' : (10.50,-4.70)
}

M.right_pos = {
 'l' : (10.50,-5.60),
 'm' : ( 8.80,-2.50),
 'n' : (12.20,-2.50),
 'o' : (11.30,-3.20),
 'p' : ( 9.70,-3.20)
}

X = [('a','p'),('c','l'),('d','n')]

fig, ax = plt.subplots(1,3,figsize=(15,5))
for i in range(3):
    ax[i].set_aspect('equal')
    ax[i].axis('off')
M.show_board(labels=True,coloured=True,ax=ax[0])
M.show_matrix(labels=True,ax=ax[1])
M.show_graph(scale=1.5,ax=ax[2])
plt.show()

M.print_rows()
M.print_columns()


```
